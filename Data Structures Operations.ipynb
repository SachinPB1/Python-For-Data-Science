### zip() function

- The zip() function in Python is a built-in function that allows you to combine two or more iterables (such as lists, tuples, etc.)

```python
# Two lists
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]

# Use zip to combine both lists
zipped = zip(names, ages)

# Convert the result to a list to view it
zipped_list = list(zipped)
print(zipped_list)
```

```python
# Two lists
keys = ["name", "age", "city"]
values = ["Alice", 25, "New York"]

# Create a dictionary using zip
my_dict = dict(zip(keys, values))

print(my_dict)
```

### loc[ ] and iloc[ ]

- Both are used for accessing data from a DataFrame


- **Label-based indexing:**  loc[] uses labels (row and column names) to access data.

```python
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
df = pd.DataFrame(data, index=['a', 'b', 'c'])


# Access data by label using loc

print(df.loc['a', 'Age'])         # (row 'a', column 'Age')
```

- **Position-based indexing:**  iloc[] uses integer-based positions (row and column index positions) to access data.

```python
# Access data by position using iloc
print(df.iloc[0, 1])               #(first row, second column)
```

### map() and apply()

- Both functions used to perform operations on a Series or DataFrame. They are often used for element-wise transformations

- **map()** is primarily used for element-wise transformations on a Series (i.e., one-dimensional data).


- It is typically used to map values from one set to another (like a lookup or replacement).

```python
# Sample data
data = pd.Series([1, 2, 3, 4])

# Map: Apply a function to each element (square the numbers)

data_1= data.map(lambda x: x*2)
print(data_1)
```

```python
# Example with Dictionary (Replacement):
data = pd.Series(['cat', 'dog', 'cat', 'rabbit'])

# Map: Replace 'cat' with 'kitten' and 'dog' with 'puppy'
replace_dict = {'cat': 'A', 'dog': 'B'}

mapped_data = data.map(replace_dict)
print(mapped_data)
```

**apply()** is a more general function that can be used on both Series and DataFrames (i.e., one-dimensional or two-dimensional data).

- It appliesa function along rows or columns of a DataFrame, or element-wise on a Series.

```python
# Sample DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Apply function to each column (default axis=0)

column_sum = df.apply(lambda x: x.sum())
print(column_sum)

# Apply function to each row (axis=1)
row_sum = df.apply(lambda x: x.sum(), axis=1)
print(row_sum)
```

```python
data = pd.Series([1, 2, 3, 4])
squared_data = data.apply(lambda x: x**2)
print(squared_data)
```
